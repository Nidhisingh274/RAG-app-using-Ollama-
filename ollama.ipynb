```python
# DATA TRANSFORMATION / DATA CHUNKING
```

```python
''' 
chunk_size should be medium if too big then more context but risk of mix data if very small chunks then high retrieval cost and more storage needed.
chunk overlap should be 10-20% of size size. If too large then can be issue of duplicates.
    '''
from langchain_text_splitters import RecursiveCharacterTextSplitter
text = RecursiveCharacterTextSplitter(chunk_size = 600, chunk_overlap = 100)
final_documents = text.split_documents(docs)
final_documents
```

```python
# Embedding
```

```python
'''
How to decide which embedding model to choose?
1. Chunk size should be inside context.
2. dimension is no. of characters in a vector of a single chunk.
3. embedding dimenions are fixed like 3072 for text-embeddings-3-large means its length is 3072.
4. bigger chunk need bigger dimension.
5. if not getting desired ouput increase dimension.
'''
```

```python
from langchain_community.embeddings import OllamaEmbeddings
embeddings = OllamaEmbeddings(model = "all-minilm:22m")             # context/token = 512
embedding_result = embeddings.embed_documents(final_documents)
embedding_result
```

```python
len(embedding_result[0]) # dimension is 384
```

```python
# Vector Store Database
```

```python
'''
How to decide which vector db to use?
When to Use	                           Recommended Vector Store	          Why
Small project (few 100–10,000 chunks)	FAISS or Chroma (local)	Easy to set up, runs on your laptop
Medium project (10,000–500,000 chunks)	Chroma, Weaviate (local)	More stable with more data
Large project (millions of chunks, production)	Pinecone, Milvus, Weaviate (cloud)	Scalable, faster search, handles huge data
'''
```

```python
from langchain_community.vectorstores import FAISS
```

```python
db = FAISS.from_documents(final_documents, embeddings)
db
```

```python
query = "Can you give some idea about transformer architecture?"
res = db.similarity_search(query)
res
```
